***Dependencies***

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

 ***Data Collection and Processing***

```python
# importimg the csv data to pandas data frame
breast_cancer_dataset = pd.read_csv('/content/data (2).csv')
```

```python
print(breast_cancer_dataset)
```

```python
# loading the data to a dataframe
data_frame=pd.DataFrame(breast_cancer_dataset)
```

```python
# prints first 5 rows
data_frame.head()
```

```python
# rows and cols in dataset
data_frame.shape
```

```python
# info aboout dataset
data_frame.info()
```

```python
# checking for missing values
data_frame.isnull().sum()
```

```python
# removing a column
del data_frame['Unnamed: 32']
```

```python
del data_frame['id']
```

```python
data_frame.info()
```

```python
# printing last 5 rows
data_frame.tail()
```

```python
data_frame['diagnosis'].value_counts()
```

```python
# statistical measures about the data
data_frame.describe()
```

***replacing data in diagnosis m with 1 and b with 0***

```python
data_frame['diagnosis']=data_frame['diagnosis'].replace(['M'],'1')
data_frame['diagnosis']=data_frame['diagnosis'].replace(['B'],'0')
```

***changing the data type in diagnosis from object to int***

```python
data_frame['diagnosis']=data_frame['diagnosis'].astype(int)
```

```python
data_frame.info()
```

```python
data_frame.shape
```

```python
data_frame.describe()
```

```python
# checking the distribution of target varibale
data_frame['diagnosis'].value_counts()
```

0= bengin
1=malignant

```python
data_frame.describe
```

```python
data_frame.groupby('diagnosis').mean()
```

 ***Separating the features and the target***

```python
X=data_frame.drop(columns='diagnosis',axis=1)
Y=data_frame['diagnosis']
```

```python
print(X)
```

```python
print(Y)
```

***splitting the data into training data and teting data***

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

***model training***
***logistic regression***

```python
model = LogisticRegression()
```

```python
# training the Logistic model using Training data

model.fit(X_train, Y_train)
```

***model evaluation***
***accuracy score***

```python
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(Y_train, X_train_prediction)
```

```python
print('Accuracy on training data =', training_data_accuracy)
```

```python
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(Y_test, X_test_prediction)
```

```python
print('Accuracy on test data =', test_data_accuracy)
```

***building a predictive system***

```python
input_data=(19.17,24.8,132.4,1123,0.0974,0.2458,0.2065,0.1118,0.2397,0.078,0.9555,3.568,11.07,116.2,0.003139,0.08297,0.0889,0.0409,0.04484,0.01284,20.96,29.94,151.7,1332,0.1037,0.3903,0.3639,0.1767,0.3176,0.1023)
# changing the input data to a numpy array
input_data_as_numpy_array=np.asarray(input_data)

#reshapping the numpy array as we are prediciting for one datapoint
input_data_reshaped=input_data_as_numpy_array.reshape(1,-1)

prediction=model.predict(input_data_reshaped)
print(prediction)
```

```python
if (prediction[0]==0):
  print('THE BREAST CANCER IS BENIGN')

else:
  print('THE BREAST CANCER IS MALIGNANT')  
```
